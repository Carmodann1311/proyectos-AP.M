ejercicio 1 f(x,y)=x+y+z=4

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D


x = 1
y = 2
z = -1


fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')


ax.scatter(x, y, z, color='r', s=100, label='(1, 2, -1)')


ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')


plt.title('Gr√°fica del punto (1, 2, -1)')


plt.legend()
plt.show()
```

ejercicio 2 f(x,y)=x+y+z=2

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D


x_val, y_val, z_val = 1, 2, -3


fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')


ax.scatter(x_val, y_val, z_val, color='red', s=100, label='(1, 2, -3)')


ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')


ax.set_title('Point (1, 2, -3) in 3D Space')


plt.show()
```
