```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def f(x, y, z):
 return x + y + 2 + 4

x = np. linspace(-10, 10, 100)
y = np. linspace(-10, 10, 100)
x, y = np.meshgrid(x, y)
z = f(x, y, 0) # Fijamos z en 0

fig = plt. figure()
ax = fig. add_subplot(111, projection= '3d')

ax. plot_surface(x, y, z, cmap='viridis')

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('f(x, y, 0)')

plt. show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

center = [0, 0, 0] # Centro de la elipsoide
radii = [3, 2, 1] # Semiejes de la elipsoide

u = np. linspace(0, 2 * np.pi, 100)
v = np. linspace(0, np.pi, 50)
x = radii[0] * np.outer(np.cos(u), np. sin(v)) + center [0]
y = radii[1] * np.outer(np. sin(u), np. sin(v)) + center [1]
z = radii[2] * np.outer(np.ones(np.size(u)), np.cos(v)) + center [2]

fig = plt. figure()
ax = fig.add_subplot(111, projection= '3d')

ax. plot_surface(x, y, z, color= 'b')

ax.set_xlim([-max(radii), max(radii)])
ax.set_ylim([-max(radii), max(radii)])
ax.set_zlim([-max(radii), max(radii)])

ax. set_xlabel ('X')
ax. set_ylabel('Y')
ax. set_zlabel('Z')

plt. show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

a = 1 # Parámetro para el eje x
b = 1 # Parámetro para el eje y
c = 1 # Parámetro para el eje z

x = np. linspace(-5, 5, 100)
y = np. linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = (x ** 2 / a ** 2) - (y ** 2 / b ** 2) + c

fig = plt. figure()
ax = fig.add_subplot(111, projection='3d')

ax. plot_surface(x, y, z, cmap='viridis')

ax.set_xlim([-5, 5])
ax.set_ylim([-5, 5])
ax.set_zlim([-10, 100])

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

plt.show()
```
